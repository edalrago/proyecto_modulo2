```python
# Importamos los módulos necesarios
import math
import numpy as np
import matplotlib.pyplot as plt

# Generamos los datos para el gráfico
x = np.array(range(-100,100))*0.1
y = np.zeros(len(x))
for i in range(len(x)):
    y[i] = math.sin(x[i])
    
def simple_sine(x, y):
    return np.sin(x) * 5.0 + np.sin(y) * 5.0

# Creamos el gráfico
plt.ion()
plt.plot(x,y)


from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

#Crear gráfica y definir tamaño
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')

#Definir ejes coordenados
xnum = np.linspace(-10,10,200)
ynum = xnum

X,Y = np.meshgrid(xnum,ynum)
Z = simple_sine(X,Y)
ax.plot_surface(X,Y,Z)
```
